**Quest-3:** What are the Combinatorial Optimization problems ?
> Combinatorial problems are used to find the an optimal solution(closed to optimal solution) from a finite collection of all possible solutions.
 
1. Navigation system in cars
2. Creating Timetable for schools
3. Decision support system in production
4. logistic enviroments like determining the ways to deliver the packages etc

Note : In such real world problems we can't apply the brute force search.

There are few Combinatorial Optimization Problems, On which we work like Travelling Salesman Problem, 

**Travelling Salesman problem**
> **Problem Statement** : Given a dataset of cities and distance between every pair of cities, the problem is to find the shortest possible route such that a salesman visits every city exactly once and returns to the starting point. 

Example-1 : Travelling Sales Man problem is an NP-Hard problem means there is no efficient algorithm that can find the optimal solution in polynomial time on classical machine.

<img src="./images/tsp1.png" width="250">

Given an undirected graph and the cost of moving between two cities is the same for each direction.

<img src="./images/Hamiltonian_path.svg" width="250">

A path through a graph that visits each vertex(city) exactly once is called a <b>hamiltonian path</b> or a <b>hamiltonian cycle</b> if it is a closed path. 


**Brute Force Method**
It considers all possible routes to determine the shortest unique solution. This method provides the optimal solution instead of an approximate, it is not feasible to use this method for problems involving a large number of cities. We needed to select the city from where we should start and at the end after covering all city we will able to reach our city from where we started our journey.

**What is `networkx` library and how to visualize the graphs using it ?**

It is a python package to create, process and visualize the graphs(network) using build-in functions.

**Problem** Graph Coloring is a NP-Hard problem for general graphs but can become NP-Completer if we ask Can graph be colored using $k$ colors for $k \geq 3$.

Graph Coloring is the procedure of assignment of colors to each vertex of a graph such that no adjacent vertices get the same color. The objective is to minimize the number of colors. The smallest number of colors required to color a graph is called the <b>chromatic number</b> of the graph.

<img src="./images/gc1.png" width="250">

```python
# visualize the graph by defining colors of the nodes as dictionary.
G = nx.Graph()
# create edges
edges = [
    ("A", "C"),
    ("A", "E"),
    ("B", "C"),
    ("B", "E"),
    ("B", "D"),
    ("C", "D"),
    ("D", "E"),
]
G.add_edges_from(edges)
# Colors the nodes
coloring = {"A": "R", "B": "R", "C": "G", "D": "B", "E": "G"}
graph_coloring_viz(G, coloring)
coloring = {"A": 0, "B": 0, "C": 1, "D": 2, "E": 1}
graph_coloring_viz(G, coloring)
```
